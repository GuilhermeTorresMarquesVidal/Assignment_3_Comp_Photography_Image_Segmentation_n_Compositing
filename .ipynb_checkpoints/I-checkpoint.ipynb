## Libraries

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.ticker import NullLocator
import matplotlib.image as mpimg
from Contents.GrabCut import rect_GrabCut
```

## Read images

```python
dados = mpimg.imread('Images/dados.png')
flor = mpimg.imread('Images/flor.png')
```

## Plotting images

```python
fig, ax = plt.subplots(figsize = (8,6))

ax.imshow(dados)

plt.gca().xaxis.set_major_locator(NullLocator())
plt.gca().yaxis.set_major_locator(NullLocator())
```

```python
fig, ax = plt.subplots(figsize = (8,6))

ax.imshow(dados)

rect = patches.Rectangle((220, 30), 250, 270, 
                         linewidth=1, 
                         edgecolor='r', 
                         facecolor='none')

ax.add_patch(rect)

plt.gca().xaxis.set_major_locator(NullLocator())
plt.gca().yaxis.set_major_locator(NullLocator())
```

```python
fig, ax = plt.subplots(figsize = (8,6))

ax.imshow(flor)

plt.gca().xaxis.set_major_locator(NullLocator())
plt.gca().yaxis.set_major_locator(NullLocator())
```

```python
fig, ax = plt.subplots(figsize = (8,6))

ax.imshow(flor)

rect = patches.Rectangle((80, 60), 350, 240, 
                         linewidth=1, 
                         edgecolor='r', 
                         facecolor='none')

ax.add_patch(rect)

plt.gca().xaxis.set_major_locator(NullLocator())
plt.gca().yaxis.set_major_locator(NullLocator())
```

## Apply grabcut

```python
rect_final = 220, 30, 250, 270

dados_cut = rect_GrabCut(
    image = (dados*255).astype(np.uint8), 
    rect_final = rect_final)
```

```python
rect_final = 80, 60, 350, 240, 

flor_cut = rect_GrabCut(
    image = (flor*255).astype(np.uint8), 
    rect_final = rect_final)
```

## Plotting image after grabcut

```python
fig, ax = plt.subplots(figsize = (8,6))

ax.imshow(dados_cut)

plt.gca().xaxis.set_major_locator(NullLocator())
plt.gca().yaxis.set_major_locator(NullLocator())

plt.savefig("Results/dados_cut.pdf", bbox_inches = 'tight')
```

```python
fig, ax = plt.subplots(figsize = (8,6))

ax.imshow(flor_cut)

plt.gca().xaxis.set_major_locator(NullLocator())
plt.gca().yaxis.set_major_locator(NullLocator())

plt.savefig("Results/flor_cut.pdf", bbox_inches = 'tight')
```
