## Libraries

```python
import numpy as np
from Contents.params import*
from Contents.read_images import read_list_of_images
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.ticker import NullLocator
```

## Read images

```python
images = read_list_of_images(
    list_of_files = list_of_files_III, 
    keys = keys_III)
```

## Plotting images

```python
fig, axs = plt.subplots(1,3, figsize = (18,6))

v = list(images.keys())
v.sort(reverse = False)

axs[0].imshow(images[v[0]])
axs[1].imshow(images[v[1]])
axs[2].imshow(images[v[2]])

titles = ['(a)\n', '(b)\n', '(c)\n']

for ax, title in zip(axs,titles):
        
        # Set title
        
        ax.set_title(title, fontsize = 24)
        
        # Remove axis
        
        ax.xaxis.set_major_locator(NullLocator())
        ax.yaxis.set_major_locator(NullLocator())
        
plt.savefig("Results/Alpha_Composition.pdf", bbox_inches = 'tight')
```

## Image composition

```python
A = images['GT04']*images['GT04_alpha']
B = images['background']*(1 - images['GT04_alpha'])
C = A + B
```

## Plotting image

```python
fig, ax = plt.subplots(figsize = (8,6))

ax.imshow(A)

plt.gca().xaxis.set_major_locator(NullLocator())
plt.gca().yaxis.set_major_locator(NullLocator())
```

```python
fig, ax = plt.subplots(figsize = (8,6))

ax.imshow(C)

plt.gca().xaxis.set_major_locator(NullLocator())
plt.gca().yaxis.set_major_locator(NullLocator())

plt.savefig("Results/image_composition.pdf", bbox_inches = 'tight')
```
