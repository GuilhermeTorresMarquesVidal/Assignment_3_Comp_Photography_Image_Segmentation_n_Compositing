## Libraries

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.ticker import NullLocator
from Contents.params import*
from Contents.masks import generate_mask
from Contents.blended import blended_pipeline
from Contents.read_images import read_list_of_images
from generate_collection import pyramid_collection, blended_collection
```

## Read images

```python
images = read_list_of_images(
    list_of_files = list_of_files_II, 
    keys = keys_II)
```

## Plotting images

```python
fig, axs = plt.subplots(1,2, figsize = (12,6))

v = list(images.keys())
v.sort(reverse = False)

axs[0].imshow(images[v[0]])
axs[1].imshow(images[v[1]])

titles = ['(a)\n', '(b)\n']

for ax, title in zip(axs,titles):
        
        # Set title
        
        ax.set_title(title, fontsize = 24)
        
        # Remove axis
        
        ax.xaxis.set_major_locator(NullLocator())
        ax.yaxis.set_major_locator(NullLocator())
        
plt.savefig("Results/Apple_Orange.pdf", bbox_inches = 'tight')
```

## Generate mask

```python
masks = dict()
masks['standing'] = generate_mask(image = images['apple'], kind = 'standing')
masks['flat'] = generate_mask(image = images['apple'], kind = 'flat')
```

## Plotting masks

```python
fig, axs = plt.subplots(1,2, figsize = (12,6))

v = list(images.keys())
v.sort(reverse = False)

axs[0].imshow(masks['standing'])
axs[1].imshow(masks['flat'])

titles = ['(a)\n', '(b)\n']

for ax, title in zip(axs,titles):
        
        # Set title
        
        ax.set_title(title, fontsize = 24)
        
        # Remove axis
        
        ax.xaxis.set_major_locator(NullLocator())
        ax.yaxis.set_major_locator(NullLocator())
        
plt.savefig("Results/masks.pdf", bbox_inches = 'tight')
```

## Generate pyramid

```python
dep_max_list = [1, 3, 5]
apple, orange, masks_col = pyramid_collection(
    images = images, masks_images = masks, dep_max_list = dep_max_list)
```

## Plotting sequence

```python
grid = plt.GridSpec(128, 126, wspace = 4.0)

axs = []
axs.append(plt.subplot(grid[:,:64]))
axs.append(plt.subplot(grid[50:,64:96]))
axs.append(plt.subplot(grid[76:,96:112]))
axs.append(plt.subplot(grid[88:,112:120]))
axs.append(plt.subplot(grid[94:,120:124]))
axs.append(plt.subplot(grid[98:,124:]))

cont = 0
for ax in axs:
    ax.imshow(apple[5]['gauss'][cont])
    ax.xaxis.set_major_locator(NullLocator())
    ax.yaxis.set_major_locator(NullLocator())
    cont = cont + 1
    
plt.savefig("Results/Apple_gauss_sequence.pdf", bbox_inches = 'tight');
```
