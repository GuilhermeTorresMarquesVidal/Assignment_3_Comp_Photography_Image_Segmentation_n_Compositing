## Libraries

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.ticker import NullLocator
from Contents.params import*
from Contents.read_images import read_list_of_images
from Contents.laplacian_pyramid import generate_pyramid
```

## Read images

```python
images = read_list_of_images(
    list_of_files = list_of_files_II, 
    keys = keys_II)
```

## Plotting images

```python
fig, axs = plt.subplots(1,2, figsize = (12,6))

v = list(images.keys())
v.sort(reverse = False)

axs[0].imshow(images[v[0]])
axs[1].imshow(images[v[1]])

titles = ['(a)\n', '(b)\n']

for ax, title in zip(axs,titles):
        
        # Set title
        
        ax.set_title(title, fontsize = 24)
        
        # Remove axis
        
        ax.xaxis.set_major_locator(NullLocator())
        ax.yaxis.set_major_locator(NullLocator())
        
plt.savefig("Results/Apple_Orange.pdf", bbox_inches = 'tight')
```

## Generate pyramid

```python
apple_gauss, apple_laplace = generate_pyramid(
    image = images['apple'], dep_max = 5)
```

```python
orange_gauss, orange_laplace = generate_pyramid(
    image = images['orange'], dep_max = 5)
```

## Plotting sequence

```python
grid = plt.GridSpec(8, 7)

ax_gauss = []
ax_gauss.append(plt.subplot(grid[:4,:4]))
ax_gauss.append(plt.subplot(grid[2:4,4:6]))
ax_gauss.append(plt.subplot(grid[3,6]))
ax_laplace = []
ax_laplace.append(plt.subplot(grid[4:,:4]))
ax_laplace.append(plt.subplot(grid[6:,4:6]))
ax_laplace.append(plt.subplot(grid[7,6]))

cont = 1
for ax in ax_gauss:
    ax.imshow(orange_gauss[-cont])
    ax.xaxis.set_major_locator(NullLocator())
    ax.yaxis.set_major_locator(NullLocator())
    cont = cont + 1
    
cont = 1
for ax in ax_laplace:
    ax.imshow(orange_laplace[-cont])
    ax.xaxis.set_major_locator(NullLocator())
    ax.yaxis.set_major_locator(NullLocator())
    cont = cont + 1
```
