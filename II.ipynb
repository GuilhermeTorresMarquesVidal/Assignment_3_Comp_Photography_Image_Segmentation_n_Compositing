## Libraries

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.ticker import NullLocator
from Contents.params import*
from Contents.masks import generate_mask
from Contents.blended import blended_pipeline
from Contents.read_images import read_list_of_images
from Contents.laplacian_pyramid import generate_pyramid
```

## Read images

```python
images = read_list_of_images(
    list_of_files = list_of_files_II, 
    keys = keys_II)
```

## Plotting images

```python
fig, axs = plt.subplots(1,2, figsize = (12,6))

v = list(images.keys())
v.sort(reverse = False)

axs[0].imshow(images[v[0]])
axs[1].imshow(images[v[1]])

titles = ['(a)\n', '(b)\n']

for ax, title in zip(axs,titles):
        
        # Set title
        
        ax.set_title(title, fontsize = 24)
        
        # Remove axis
        
        ax.xaxis.set_major_locator(NullLocator())
        ax.yaxis.set_major_locator(NullLocator())
        
plt.savefig("Results/Apple_Orange.pdf", bbox_inches = 'tight')
```

## Generate mask

```python
masks = dict()
masks['standing'] = generate_mask(image = images['apple'], kind = 'standing')
masks['flat'] = generate_mask(image = images['apple'], kind = 'flat')
```

## Plotting mask

```python
fig, ax = plt.subplots(figsize = (8,6))

ax.imshow(masks['standing'])

plt.gca().xaxis.set_major_locator(NullLocator())
plt.gca().yaxis.set_major_locator(NullLocator())

plt.savefig("Results/mask1.pdf", bbox_inches = 'tight')
```

```python
fig, ax = plt.subplots(figsize = (8,6))

ax.imshow(masks['flat'])

plt.gca().xaxis.set_major_locator(NullLocator())
plt.gca().yaxis.set_major_locator(NullLocator())

plt.savefig("Results/mask2.pdf", bbox_inches = 'tight')
```

## Generate pyramid

```python
apple_gauss, apple_laplace = generate_pyramid(
    image = images['apple'], dep_max = 5)
```

```python
orange_gauss, orange_laplace = generate_pyramid(
    image = images['orange'], dep_max = 5)
```

```python
mask_standing_gauss, mask_standing_laplace = generate_pyramid(
    image = masks['standing'], dep_max = 5)
```

```python
mask_flat_gauss, mask_flat_laplace = generate_pyramid(
    image = masks['flat'], dep_max = 5)
```
